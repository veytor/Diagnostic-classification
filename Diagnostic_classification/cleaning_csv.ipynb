# ESTANCIAS HOSPITALES

```python
df_estancias = pd.read_csv("Data/Estancia_comunidades.csv", sep = "\t")

df_estancias
```

```python
#df_estancias.loc[df_estancias["Total"]=="0"] 
```

```python
df_estancias["Total"] = df_estancias["Total"].str.replace(".","").astype(int) # Especifico que la columna total es tipo número
```

```python
df_estancias.info()
```

```python
df_estancias.groupby("Comunidades y Ciudades Autónomas").sum() # Realizo un groupby 
```

```python
df_estancias["Comunidades y Ciudades Autónomas"] = df_estancias["Comunidades y Ciudades Autónomas"].str.capitalize()
```

```python
df_estancias = df_estancias.loc[df_estancias["Sexo"] == "Ambos sexos"]
```

```python
df_estancias.loc[df_estancias["Provincias"].isnull()]
```

```python
mierda = df_estancias.loc[~df_estancias["Provincias"].isnull()]

mierda
```

```python
mierda.groupby(["Diagnóstico principal (CIE10MC)", "Comunidades y Ciudades Autónomas", "Provincias"]).sum()
```

```python
mierda.groupby(["Diagnóstico principal (CIE10MC)", "Comunidades y Ciudades Autónomas", "Provincias"]).sum().loc[("TODAS LAS CAUSAS A00-Z99 (salvo los códigos V, W, X, Y)", "Andalucía")].sum()
```

```python
df_estancias = df_estancias.loc[df_estancias["Provincias"].isnull()]
```

```python
df_estancias = df_estancias.drop(columns=["Sexo", "Provincias"])
```

```python
df_estancias
```

```python
df_total_nacional = df_estancias.loc[(df_estancias["Total Nacional"]=="TOTAL NACIONAL") & ~(df_estancias["Comunidades y Ciudades Autónomas"].isnull())]
```

```python
df_total_extranjero = df_estancias.loc[df_estancias["Total Nacional"]=="Extranjero"]
```

```python
df_total_nacional
```

```python
df_total_extranjero
```

```python
df_estancias = pd.concat([df_total_nacional, df_total_extranjero]).reset_index(drop = True)
```

```python
df_estancias
```

```python
df_estancias.info()
```

```python
df_estancias.to_csv("Data/clean_estancias.csv", index= False) # Guardo el archivo en CSV sin índices
```
